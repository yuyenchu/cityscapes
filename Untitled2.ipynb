```python
from clearml import Dataset
import tensorflow as tf
dataset = Dataset.get(dataset_project='semantic_segmentation', dataset_name='cityscapes_fine', dataset_version='1.0.0-c416')
dataset_path = dataset.get_local_copy()

def load_image(data_path):
    # img_path = tf.strings.regex_replace(data_path, '_gtFine_color', '_leftImg8bit')
    img_path = tf.strings.regex_replace(data_path, '_gtFine_labelIds', '_leftImg8bit')
    
    mask_path = data_path
    
    image = tf.io.read_file(img_path)
    input_image = tf.image.decode_png(image, channels=3)
#     input_image = tf.image.resize(input_image, (416, 416))
    input_image = tf.cast(input_image, tf.float32) / 255.0
    
    mask = tf.io.read_file(mask_path)
    input_mask = tf.image.decode_png(mask)
    # input_mask = tf.cast(mask, tf.int32)
    return input_image, input_mask
class Augment(tf.keras.layers.Layer):
    def __init__(self, seed=42):
        super().__init__()
        # both use the same seed, so they'll make the same random changes.
        self.augment_inputs = tf.keras.layers.RandomFlip(mode="horizontal", seed=seed)
        self.augment_labels = tf.keras.layers.RandomFlip(mode="horizontal", seed=seed)

    def call(self, inputs, labels):
        inputs = self.augment_inputs(inputs)
        labels = self.augment_labels(labels)
        return inputs, labels
train_images = tf.data.Dataset.list_files(f'{dataset_path}/train/*/*_gtFine_labelIds.png').map(load_image, num_parallel_calls=tf.data.AUTOTUNE)
train_batches = train_images.map(Augment())
```

```python
%matplotlib inline
from clearml import Dataset
from utils import load_image, Augment
import matplotlib.pyplot as plt
dataset = Dataset.get(dataset_project='semantic_segmentation', dataset_name='cityscapes_fine', dataset_version='1.0.0-c416')
dataset_path = dataset.get_local_copy()
print(dataset_path)
img_batches = tf.data.Dataset\
.list_files(f'{dataset_path}/val/*/*_gtFine_labelIds.png')\
.map(load_image, num_parallel_calls=tf.data.AUTOTUNE)\
.batch(12)\
.repeat().map(Augment())

def display(display_list):
    plt.figure(figsize=(15, 15))

    title = ['Input Image', 'True Mask', 'Predicted Mask']

    for i in range(len(display_list)):
        plt.subplot(1, len(display_list), i+1)
        plt.title(title[i])
        plt.imshow(tf.keras.utils.array_to_img(display_list[i]))
        plt.axis('off')
    plt.show()
    
for images, masks in img_batches.take(5):
    sample_image, sample_mask = images, masks
    print(sample_image.shape,sample_mask.shape)
    display([sample_image, sample_mask])
    
def show_predictions(model, dataset=None, num=1):
    if dataset:
        for image, mask in dataset.take(num):
            pred_mask = model.predict(image)
            display([image[0], mask[0], create_mask(pred_mask)])
    else:
        display([sample_image, sample_mask])
```
